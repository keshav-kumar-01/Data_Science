```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=6*np.random.rand(100,1)-3
y=0.5*X**2+1.5*X+2+np.random.randn(100,1)
# quadratic equation used y = 0.5x^2 + 1 .5x+2+outliers
plt.scatter(X,y)
plt.xlabel('X_dataset')
plt.ylabel('Y_dataset')
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test =train_test_split(X,y,test_size=0.25 , random_state=42)
```

```python
## let's implement a simple linear regression 
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
X_train
```

```python
regressor.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
score1= r2_score(y_test,regressor.predict(X_test))
print(score1)
```

```python
## lets vizualize this model
plt.plot(X_train,regressor.predict(X_train),color='r')
plt.scatter(X_train , y_train)
plt.xlabel("X_dataset")
plt.ylabel("y_dataset")
```

```python
#lets apply ploynomial transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=4,include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
from sklearn.metrics import r2_score
regressor = LinearRegression()
regressor.fit(X_train_poly,y_train)
y_pred=regressor.predict(X_test_poly)
score = r2_score(y_test,y_pred)
print(score)
```

```python
print(regressor.coef_)
```

```python
print(regressor.intercept_)
```

```python
plt.scatter(X_train,regressor.predict(X_train_poly))
plt.scatter(X_train,y_train)
```

```python
#pipeline polynomical 
```

```python
from sklearn.pipeline  import Pipeline
```

```python
def poly_pipeline(degree):
    X_new = np.linspace(-3,3,200).reshape(200,1)
    poly_features=PolynomialFeatures(degree = degree , include_bias=True)
    lin_reg=LinearRegression()
    poly_regression = Pipeline([
        ("poly_features",poly_features),
        ("lin_reg",lin_reg)
    ])
    poly_regression.fit(X_train,y_train)  # polynomial and fit for linear regression
    y_pred_new=poly_regression.predict(X_new)
    #plotting prediction line
    plt.plot(X_new,y_pred_new,'r',label="degree"+str(degree),linewidth=2)
    plt.plot(X_train,y_train,"b.",linewidth=3)
    plt.plot(X_test,y_test,"g.",linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis([-4,4,0,10])
    plt.show()
```

```python
poly_pipeline(6)
```
