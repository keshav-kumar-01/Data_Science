```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.30)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
```

```python
logistic.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
#hyper parameter tuning and cross validation
model = LogisticRegression()
penalty = ['l1','l2','elasticnet']
c_values = [100 , 10 , 1.0 , 0.1 , 0.01]
solvers=['newton-cg','lbgfs','liblinear','sag','saga']
```

```python
params = [{
    'solver':['liblinear'],
    'penalty':['l1','l2'],
    'C':[0.01,1.0,1.0,10.0]
},
{
    'solver':['newton-cg'],
    'penalty':['l2'],
    'C':[0.01,0.1,1.0,10.0]
}]
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
# GRID  SEARCH CV
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)

```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
#Randomsize CV
from sklearn.model_selection import RandomizedSearchCV
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5, scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred=randomcv.predict(X_test)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
X,y= make_classification(n_samples=1000,n_classes=2,random_state=1)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,random_state=42,test_size=0.25)
```

```python
dummy_model_prob = [0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
#prediction based on the proability
model_prob=  model.predict_proba(X_test)
```

```python
#let's focus on the positive
model_prob=model_prob[:,1]
```

```python
dummy_model_auc=roc_auc_score(y_test,dummy_model_prob)
model_auc = roc_auc_score(y_test,model_prob)
print(dummy_model_auc)
print(model_auc)
```

```python
#calculate ROC curve
dummy_fpr,dummy_tpr,_=roc_curve(y_test,dummy_model_prob)
model_fpr,model_tpr,thresholds =roc_curve(y_test,model_prob)
```

```python
thresholds
```

```python
plt.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
plt.plot(model_fpr,model_tpr,linestyle=':',label='Logistic')
plt.xlabel('False Positve Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20, 50))
ax = fig.add_subplot(111)

ax.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy model')
ax.plot(model_fpr, model_tpr, linestyle=':', label='Logistic')

for i, (x, y, t) in enumerate(zip(model_fpr, model_tpr, thresholds)):
    if i % 5 == 0:
        ax.annotate(f'{t:.2f}', (x, y))

ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.legend()
plt.show()
```
