```python
##basic plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
### create a scatterplot
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("scatterplot for total bill and tips")
```

```python
##line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Total bill by size')
```

```python
### categorical plots
## bar
sns.barplot(x = 'day',y='total_bill',data=tips)
```

```python
## box plot (used for outliers)
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
##violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
### histogram 
sns.histplot(tips['total_bill'],bins=10, kde=True)## kde = is for curve(full form of kde is kernal density estimator)
```

```python
###KDE plots
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
### pair plot 
sns.pairplot(tips)
```

```python
#heatmap
corr=tips[['total_bill','tip','size']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
