```python
import pandas as pd

def isduplicate(file_path, subset_columns=None):
    """
    Check if rows in an Excel/CSV file are duplicates.

    Parameters:
    file_path (str): Path to the Excel or CSV file.
    subset_columns (list, optional): List of column names to check for duplicates.
                                      If None, all columns are used.

    Returns:
    pd.DataFrame: Original data with a new column 'is_duplicate' (True/False).
    """
    # Load the file
    if file_path.endswith('.xlsx'):
        df = pd.read_excel(file_path)
    else:
        df = pd.read_csv(file_path)
    
    # Create a boolean column marking duplicates
    df['is_duplicate'] = df.duplicated(subset=subset_columns, keep=False)
    
    return df

# Example usage
file_path = "Copy of hudle_court_google_maps_enriched-_2_(1).xlsx"  # or "data.csv"
result_df = isduplicate(file_path, subset_columns=["Name","Location","Description","Rating","Highlights","Latitude_Longitude","Image_URLs","Court_URL","Phone"])
print(result_df)

# Save to new file (optional)
result_df.to_excel("output_with_duplicates.xlsx", index=False)
```

```python
import pandas as pd
import os

def remove_duplicates(file_path, subset_columns=None):
    """
    Load Excel/CSV file, remove duplicates, and return cleaned dataframe.
    
    Parameters:
        file_path (str): Path to Excel or CSV file.
        subset_columns (list): Columns to check for duplicates. If None, uses all columns.
        
    Returns:
        pd.DataFrame: Data without duplicates.
    """
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    # Load the file
    if file_path.lower().endswith(('.xlsx', '.xls')):
        df = pd.read_excel(file_path)
    else:
        df = pd.read_csv(file_path)

    print("Original row count:", len(df))

    # Remove duplicates
    df_clean = df.drop_duplicates(subset=subset_columns, keep="first")

    print("Row count after removing duplicates:", len(df_clean))
    print(f"Removed {len(df) - len(df_clean)} duplicate rows.")

    return df_clean

# ---------------------------
# Example usage
# ---------------------------
file_path = "Copy of hudle_court_google_maps_enriched-_2_(1).xlsx"

# Make sure these match EXACTLY your Excel column names
subset_columns = [
    "Name",
    "Location",
    "Description",
    "Rating",
    "Highlights",
    "Latitude_Longitude",
    "Image_URLs",
    "Court_URL",
    "Phone"
]

df_no_duplicates = remove_duplicates(file_path, subset_columns=subset_columns)

# Save the cleaned file
df_no_duplicates.to_excel("cleaned_file.xlsx", index=False)
print("Cleaned file saved as cleaned_file.xlsx")
```
