```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
## create a line plot
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('basic plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
##create cosutomized line plot
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=2,markersize=5)
plt.grid(True)
```

```python
### multiple plot
## sample data
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]


plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("plot1")

plt.subplot(1,2,2)
plt.plot(y1,x,color='green')
plt.title("plot1")
```

```python
#bar plot
categories=['A','B','C','D','E']
values=[5,7,3,8,6]
#create a bar plot
plt.bar(categories,values,color='purple')
```

```python
# histogram 
#histogram are used to represent the distribution of dataset , they divide the data into bins and count the numeber of data points in each bin
```

```python
data=[1,2,3,4,5,2,2,3,2,1,4,5,3,2,3,1,2,4,5,4,3,2,1]
```

```python
##create a histogram
plt.hist(data,bins=5)
```

```python
## create a scatter plot
x=[1,2,3,4,5]
y=[2,3,4,5,6]
plt.scatter(x,y)
```

```python
##pie chart
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)

#create a pie chart
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%")
```

```python
import pandas as pd
```

```python
sales_data_df = pd.read_csv('Sales_data.csv')
```

```python
sales_data_df.head()
```

```python
## plot total sales by products
sales_data_df.info()
```

```python
total_sales_by_product=sales_data_df.groupby('Product Category')['Total Revenue'].sum()
```

```python
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
#plot sales trend over time
sales_trend=sales_data_df.groupby('Date')['Total Revenue'].sum().reset_index()
print(sales_trend)
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
