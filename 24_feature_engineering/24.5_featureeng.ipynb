```python
# nominal encoding 
#one hot encoding also known as nominal encoding , is a technique used to represent categorical data as numerical data which is 
#more suitable for ML alogorithms . In this techniques each category is represented as a binary vector where each bit corresponds 
#to a unique category . For Example if we have a categorical variable "color" with three posiblilities (red , blue , green ) we can 
#represent it on One Hot Encoding as follows 

#1 Red :[1,0,0]
#2 Green : [0,1,0]
#3 Blue : [0,0,1]
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
### create a  simple data frame
df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
### create an instance of OneHotEncoder
encoder=OneHotEncoder()
```

```python
## perform fit and then transform
encodeed=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
```

```python
encoder_df = pd.DataFrame(encodeed,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
###label encoding
# label encoding and ordinal encoding are two techniques used to encode categorical data as numerical data
#label encoding involves assigning a unique numerical label to each category in the variable the labels are usually assigned in 
#alphabatical order or based on the frequency of the categories for example if we have a categorical variable "color" with three 
#possible value (red,green,blue)we can represent it using label encoding as follows :
#1 . red - 1 
#2. green - 2
#3 . blue - 3
```

```python

### create a  simple data frame
df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

```python
lbl_encoder.transform([['blue']])
```

```python
lbl_encoder.transform([['green']])
```

```python
###ordinal encoding
# It is used to encode categorical data that have an intrinsic order or ranking , in this technique each cateogory is assigned a
# numerical value based on its position in the order . For example if we have a categorical variable "education level" with four
# possible values (high school , college , graduate , post graduate ), we can represent it using ordinal encoding as follows :
# 1. high school :1
# 2. college :2
# 3. graduate :3
# 4. Post graduate:4
```

```python
# ordinal encoding
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame(
    {
        'size' : ['small','medium','large','small','medium','large']
    }
)
```

```python
df
```

```python
### create an instance then fit_transform
encoder = OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

```python
encoder.transform([['large']])
```

```python
## Target Guided Ordinal Encoding
#It is a technique used to encode categorical variables based on their relationship with the target variable . This encoding technique
#is useful when we have a categorical variable with a large number of unique categories and we want to use this variable as a
#feature in our machine learning model .
# in target guided ordinal encoding we replace each category in categorical variable with a numerical values based on the mean or 
# median of the target variable for that category this creates a monotonic relationship between the categorical variable and the 
# target variable which can improve the predictive power of our model . 
```

```python
import pandas as pd
df = pd.DataFrame(
    {
        'city':['New York','London','Paris','Tokyo','New York','Paris'],
        'price':[200,150,300,250,180,320]
    }
)
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
df[['price','city_encoded']]
```
