```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
## convert the reviews datatype in int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy['Size'].unique()
```

```python
#size 19M -> 19000000 --> 19000k
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy['Size']
```

```python
df_copy['Installs'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for items in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(items,'')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype(int)
df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
## handlling last update feature
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.to_csv('playstore_cleaned.csv')
```

```python
df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')
```

```python
df_copy.shape
```

```python
numerical_feature = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_feature = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

print('We have {} numerical features : {}'.format(len(numerical_feature), numerical_feature))
print('We have {} categorical features : {}'.format(len(categorical_feature), categorical_feature))
```

```python
## what is the propotion of count on categorial columns
for col in categorical_feature:
    print(df[col].value_counts(normalize=True)*100)
    print('---------------------------------------')
```

```python
## vizualization of propotion of count data on categorical columns
plt.figure(figsize=(15,15))
plt.suptitle('Univarient analysis of numerical features', fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(0, len(numerical_feature)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[numerical_feature[i]],shade=True,color='r')
    plt.xlabel(numerical_feature[i])
    plt.tight_layout()
```

```python
## vizualization of propotion of count data on categorical columns
plt.figure(figsize=(15,15))
plt.suptitle('Univarient analysis of categorical features', fontsize=20,fontweight='bold',alpha=0.8,y=1.)

category=['Type','Content Rating']
for i in range(0, len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
### which is the most popular app category ?
df_copy['Category'].value_counts().plot.pie(y=df['Category'],figsize=(15,16),autopct='%1.1f')
```
