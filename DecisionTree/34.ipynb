```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
#Independent feature
X=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
#dependent features
y = iris['target']
```

```python
# train test split 
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.20 , random_state=10)
```

```python
# decision tree classifier

from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
#vizualize the decision Tree
from sklearn import tree
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report 
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

```python
params = { 
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' :['best','random'],
    'max_depth' :[1,2,3,4,5],
    'max_features' : ['auto','sqrt','log2']

}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treemodel,param_grid=params,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
grid.best_score_
```

```python
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```
